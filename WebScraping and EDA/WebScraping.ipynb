<h1> Advanced WebScraping with Beautiful Soup and Requests </h1>


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

Scraping the tags of html through online

```python
url = 'https://en.wikipedia.org/wiki/Web_scraping'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

# Extract all headings (h1, h2, h3,)
for tag in soup.find_all(['h1', 'h2', 'h3']):
    print(tag.text.strip())
```

Scraping the tables from online

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
tables = pd.read_html(url)  # Extract all tables

# Print the third table
df = tables[2]
print(df.head())  # first 5 rows
```

Scraping the href hyperlinks through online

```python
url = 'https://www.python.org'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract all anchor tags
for link in soup.find_all('a', href=True):
    print(link['href'])
```

Scraping the paginated(pages) content through online

```python
# https://quotes.toscrape.com/page/1/ open this page to get the first page
base_url = 'https://quotes.toscrape.com/page/{}/' # {} used this for to get all pages
for page in range(1, 4):     # to get the first three pages
  url = base_url.format(page)
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  quotes = soup.find_all('span', class_='text')  # Finds all quote texts and their corresponding authors using HTML class names.
  authors = soup.find_all('small', class_='author')
  for quote, author in zip(quotes, authors):   # Pairs each quote with its author and stores them as dictionaries in a list.
    print(quote.text, '-', author.text)
```
